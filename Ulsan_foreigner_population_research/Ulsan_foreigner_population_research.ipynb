```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
plt.rc('font', family='New Gulim')
```

```python
df = pd.read_csv('Ulsan_foreigner_population.csv', encoding='cp949')
df = df.drop(columns=['기준일자'])

country = []
p = []

for i in range(len(df)):
    if df['국적'][i] not in country:
        country.append(df['국적'][i])

sum = 0
for i in country:
    for j in range(len(df)):
        if i == df['국적'][j]:
            sum += int(df[' 인구수 '][j])
    p.append(sum)
    sum = 0

plt.figure(figsize=(15, 15))
plt.subplot(211)

plt.bar(np.arange(len(country)), p, width = 0.8, color="skyblue")
plt.xticks(np.arange(len(country)), country)
plt.xlabel('나라별', size=15)
plt.ylabel('외국인 인구수', size=15)
plt.title('울산시 나라별 외국인 인구수')
for i in range(len(country)):
    plt.text(i, p[i], p[i], 
             horizontalalignment='center',
             verticalalignment='bottom')
    
plt.subplot(212)
explode = [0, 0, 0.1, 0.2, 0, 0, 0] 
wedgeprops={'width': 0.7, 'edgecolor':'w', 'linewidth':5}
plt.pie(p, labels=country, autopct='%.1f%%', explode=explode, counterclock=False, shadow=False, wedgeprops=wedgeprops)
plt.show()
```
